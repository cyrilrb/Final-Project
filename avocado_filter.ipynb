```python
# Dependencies and Setup
import pandas as pd

```

```python
# File to Load (Remember to Change These)
file_to_load = "avocado.csv"

# Read Purchasing File and store into Pandas data frame
avocado_data = pd.read_csv(file_to_load)
# avocado_data.head()
avocado_data.loc[avocado_data['region'] == 'Portland']
```

```python
avocado_data['region'].replace({
'BaltimoreWashington': 'Baltimore',
'BuffaloRochester':'Buffalo',
'CincinnatiDayton':'Cincinnati',
'DallasFtWorth':'Dallas',
'HarrisburgScranton':'Harrisburg',
'HartfordSpringfield':'Hartford',
'MiamiFtLauderdale':'Miami',
'NewOrleansMobile':'New Orleans',
'NewYork':'New York',
'LosAngeles':'Los Angeles',
'PhoenixTucson':'Phoenix',
'RaleighGreensboro':'Raleigh',
'RichmondNorfolk':'Richmond',
"SanDiego":"San Diego",
'SanFrancisco':'San Francisco',
'SouthCarolina':'South Carolina',
"StLouis":"St. Louis"}, inplace = True
)


# avocado_data
avocado_data.loc[avocado_data['region'] == 'Portland']
```

```python
file_to_load2 = "worldcities.csv"

# Read Purchasing File and store into Pandas data frame
world_data = pd.read_csv(file_to_load2)
world_data = world_data[world_data['population'] > 100000]
world_data.head()

# world_data.loc[world_data['city'] == 'Portland']
```

```python
half_world = world_data.loc[:,['city', 'admin_name', 'country', 'lat', 'lng']]
half_world = half_world.loc[half_world['country'].isin(['United States'])]
half_world.head()
```

```python
# Rename the 'city' column to 'region' so that we can merge the data
half_world.rename(columns = {'city':'region',
'admin_name': 'state',
}, inplace = True)

half_world
```

```python
avocado_world = pd.merge(avocado_data, half_world, on='region')
avocado_world.head(2)
```

```python
avocado_world['region'].isnull().values.any()
```

```python
avocado_world = avocado_world[ (avocado_world.state != 'Maine')]

avocado_world.loc[avocado_world['region'] == 'Portland']
```

```python
avocado_world.to_csv('avocado_world.csv', index = False)
```
